## Price Distribution

This visualization illustrates the distribution of flight ticket prices in the dataset. It helps identify common price ranges, the spread of ticket prices, and the presence of any potential outliers.

```python
plt.figure(figsize=(10,6))

sns.histplot(df['Price'], bins=30, kde=True)

plt.axvline(df['Price'].mean(), linestyle='--', label='Mean')
plt.axvline(df['Price'].median(), linestyle='-', label='Median')

plt.title('Price Distribution')
plt.xlabel('Ticket Price')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

From the histogram of flight ticket prices, we observe that the distribution is **right-skewed (positively skewed)**. This indicates that most flight tickets are priced in the lower to mid range, while a smaller number of tickets are significantly more expensive.

The **mean is higher than the median**, which further confirms the presence of high-value outliers pulling the average upward. This suggests that a few premium or long-distance flights are influencing the overall price distribution.

Overall, the majority of ticket prices are concentrated around the lower price range, while only a few flights fall into the high-price segment.
